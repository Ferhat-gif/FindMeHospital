```python
import pandas as pd
df = pd.read_csv('C:/Users/48517/Desktop/kurslar/FindMeHospital/FindMeHospitalv1/csvToExcel/Zeszyt1-K.csv', delimiter=';')

```

```python
df['Insurance Type'].fillna('None', inplace=True)
```

```python
df.head()
```

```python
df['Symptoms'].fillna('None', inplace=True)
df['Symptom Duration'].fillna('None', inplace=True)
```

```python
df.head()
```

```python
df['Phone'].head(5).tolist()
```

```python
df['Gender'] = df['Gender'].str.strip()
df['First Name'] = df['First Name'].str.strip()
df['Last Name'] = df['Last Name'].str.strip()
df['Location'] = df['Location'].str.strip()
df['Has Insurance'] = df['Has Insurance'].str.strip()
df['Insurance Type'] = df['Insurance Type'].str.strip()
df['Complaint'] = df['Complaint'].str.strip()
df['Symptoms'] = df['Symptoms'].str.strip()
df['Symptom Duration'] = df['Symptom Duration'].str.strip()
```

```python
df.head()
```

```python
df['Complaint'] = df['Complaint'].str.title()
df['Symptoms'] = df['Symptoms'].str.lower()
```

```python
df.head()
```

```python
df[['Gender', 'Symptom Duration']].head(10)
```

```python
df.columns
```

```python
df['Gender'].unique()
```

```python
df['Height (cm)'].dtype
```

```python
df['Height (cm)'] = pd.to_numeric(df['Height (cm)'], errors='coerce')
```

```python
df['Height (cm)'].dtype
```

```python
df['BMI'] = df['Weight (kg)'] / ((df['Height (cm)'] / 100) ** 2)
```

```python
bins = [0, 18, 35, 50, 65, 100]
labels = ['Child', 'Young Adult', 'Adult', 'Middle Age', 'Senior']
df['Age Group'] = pd.cut(df['Age'], bins=bins, labels=labels)
```

```python
from collections import Counter

# 1. Drop missing values (if any)
complaint = df['Complaint'].dropna()

# 2. Split symptoms by comma, clean each part, and flatten into one list
all_complaint = complaint.str.lower().str.strip().str.split(',')  # split by comma
all_complaint = [s.strip() for sublist in all_complaint for s in sublist]  # flatten and strip

# 3. Count occurrences
complaint_counts = Counter(all_complaint)
complaint_counts


```

```python
# 4. Convert to Series and plot top 10
pd.Series(complaint_counts).sort_values(ascending=False).head(10).plot(kind='barh')
plt.title('Top 10 Reported Complaints')
plt.xlabel('Number of Patients')
plt.ylabel('Complaint')
plt.show()
```

```python
# Filter only Young Adults
young_adults = df[df['Age Group'] == 'Young Adult']

# Count complaints in that group
common_complaints = young_adults['Complaint'].value_counts().head(5)

# Plot
common_complaints.plot(kind='barh')
plt.title('Top Complaints Among 18-35')
plt.xlabel('Number of Patients')
plt.ylabel('Complaint')
plt.show()
```

```python
from collections import Counter

# 1. Filter only Young Adults
youngAdults_complaint = df[df['Age Group'] == 'Young Adult']

# 2. Drop NaNs and split the symptoms
Complaints = youngAdults_complaint['Complaint']
split_complaint = Complaints.str.lower().str.strip().str.split(',')

# 3. Flatten and clean
flat_complaint = [s.strip() for sublist in split_complaint for s in sublist]

# 4. Count
complaint_counts = Counter(flat_complaint)

# 5. Optional: Plot
import matplotlib.pyplot as plt
import pandas as pd

pd.Series(complaint_counts).sort_values(ascending=False).head(10).plot(kind='barh')
plt.title('Top Complaints Among 18-35 Ages')
plt.xlabel('Number of Patients')
plt.ylabel('Complaint')
plt.show()
```
